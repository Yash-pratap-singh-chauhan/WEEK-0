```python
# --- 3. Data Exploration ---
print("Shape:", df.shape)
print(df.info())
print(df.describe())
```

```python
# --- 4. Visualization ---

# 4.1 Histograms
plt.figure(figsize=(18, 5))
for i, col in enumerate(['Age', 'Annual Income (k$)', 'Spending Score (1-100)']):
    plt.subplot(1, 3, i+1)
    sns.histplot(df[col], kde=True, color=['blue', 'green', 'red'][i])
    plt.title(f"{col} Distribution")
    plt.xlabel(col)
    plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# 4.2 Boxplots
plt.figure(figsize=(18, 5))
for i, col in enumerate(['Age', 'Annual Income (k$)', 'Spending Score (1-100)']):
    plt.subplot(1, 3, i+1)
    sns.boxplot(x=df[col], color=['blue', 'green', 'red'][i])
    plt.title(f"{col} Boxplot")
plt.tight_layout()
plt.show()

# 4.3 Scatterplot: Annual Income vs Spending Score
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=df, s=80)
plt.title('Annual Income vs Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()

# 4.4 Pairplot
sns.pairplot(df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)', 'Gender']], hue='Gender')
plt.suptitle('Pairplot of Features', y=1.02)
plt.show()
```

```python
# --- 5. Skewness and Normalization ---

# 5.1 Skewness before transformation
for col in ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']:
    print(f"{col} skewness: {skew(df[col]):.2f}")

# 5.2 Transformations
df['Log Income'] = np.log(df['Annual Income (k$)'] + 1)
scaler_standard = StandardScaler()
scaler_minmax = MinMaxScaler()
df['Standard Income'] = scaler_standard.fit_transform(df[['Annual Income (k$)']])
df['MinMax Income'] = scaler_minmax.fit_transform(df[['Annual Income (k$)']])

# 5.3 Skewness after transformation
print("\nAfter log transform:", skew(df['Log Income']))
print("After StandardScaler:", skew(df['Standard Income']))
print("After MinMaxScaler:", skew(df['MinMax Income']))

# 5.4 Plot transformed income
plt.figure(figsize=(18, 5))
for i, col in enumerate(['Log Income', 'Standard Income', 'MinMax Income']):
    plt.subplot(1, 3, i+1)
    sns.histplot(df[col], kde=True, color=['purple', 'orange', 'teal'][i])
    plt.title(f"{col} Distribution")
plt.tight_layout()
plt.show()
```

```python
# --- 6. K-Means Clustering ---

# 6.1 Prepare data for clustering
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# 6.2 Fit KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)
   
# 6.3 Plot clusters
plt.figure(figsize=(10, 7))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', 
                hue='Cluster', palette='viridis', data=df, s=100)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], 
            s=300, c='red', marker='X', label='Centroids')
plt.title('K-Means Clustering (3 Clusters)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()

# 6.4 Show cluster centers
print("Cluster Centers (Annual Income, Spending Score):")
print(kmeans.cluster_centers_)
```
